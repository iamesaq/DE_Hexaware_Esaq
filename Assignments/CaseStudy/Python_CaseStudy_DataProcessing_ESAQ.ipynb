##### Task 6: Missing Data Handling

Checking for missing values

```python
# prints sum of null values in each column
df.isnull().sum()
```

```python
df.head(24)
```

Filling missing values in LoanAmount with mean value

```python
loan_amount_mean = df['LoanAmount'].mean()
df['LoanAmount'].fillna(loan_amount_mean, inplace=True)
```

Filling missing values in Gender using mode(most frequent value)

```python
gender_mode = df['Gender'].mode()[0]
df['Gender'].fillna(gender_mode, inplace=True)
```

Filling departments, self_employed, credit_history columns missing with Mode

```python
for col in ['Dependents', 'Self_Employed', 'Credit_History','Married']:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mode()[0], inplace=True)
```

```python
df.head(24)
```

Checking for missing values again

```python
df.isnull().sum()
```

##### Task 7: Sorting the dataframe by ApplicantIncome

##### Task 8: Merging Dataframes 

```python
# create a second df
loan_status_df = pd.DataFrame({
    'Loan_ID': ['LP001002', 'LP001003', 'LP001005'],
    'Loan_Status_Detail': ['Approved', 'Rejected', 'Approved']
})

# merge the two df
merged_df = pd.merge(df, loan_status_df, on='Loan_ID', how='left')
merged_df.head()
```

##### Task 9: Applying Functions in Dataframe

```python
def catorgorize_income(income):
    if income < 5400:
        return "Low"
    else:
        return "High"
    
df["IncomeCategory"] = df["ApplicantIncome"].apply(catorgorize_income)
df[['Loan_ID', 'ApplicantIncome', 'IncomeCategory']].head()
```

##### Task 10: Lambda function in Dataframe

```python
df['Total_Income'] = df.apply(lambda row: row['ApplicantIncome'] + row['CoapplicantIncome'], axis=1)
print(df[['Loan_ID', 'ApplicantIncome', 'CoapplicantIncome', 'Total_Income']].head())
```

##### Task 11: Visualizing the Dataframe

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(df['ApplicantIncome'], bins=50, color='skyblue', edgecolor='black')
plt.title('Distribution of Applicant Income')
plt.xlabel('Applicant Income')
plt.ylabel('Frequency')
plt.grid(True)
plt.savefig('applicant_income_histogram.png')

plt.figure(figsize=(8, 5))
df['Loan_Status'].value_counts().plot(kind='bar', color=['lightgreen', 'salmon'])
plt.title('Loan Status Counts')
plt.xlabel('Loan Status')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.savefig('loan_status_barchart.png')
```
